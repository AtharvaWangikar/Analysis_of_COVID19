```python
df =pd.read_csv(r'C:/Users/athar/OneDrive/Documents/GitHub/Analysis_of_COVID19/covid_19_india.csv')
```

```python
df.head()
```

```python
#i want to rename column name (State/UnionTerritory,Cured)

df =df.rename(columns={'State/UnionTerritory':'States','Cured':'Recovery'})
```

```python
df.head()
```

```python
# know i need to drop some  unnecessary columns(Sno,ConfirmedIndianNational,ConfirmedForeignNational,Time)

df = df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],axis=1)
```

```python
df
```

```python
#now i want to check Active cases

df['Active'] =df['Confirmed']- df['Recovery']- df['Deaths']
```

```python
df.head()
```

```python
data =df.sort_values(['Date','States']).reset_index(drop=True)
data['Date'] =pd.to_datetime(data['Date'])
```

```python
data
```

```python
#finding missing values

india_cases = data[data['Date']==data['Date'].max()].copy().fillna(0)
india_cases.index =india_cases['States']

india_cases =india_cases.drop(['States','Date'],axis=1)
```

```python
df =pd.DataFrame(pd.to_numeric(india_cases.sum()),dtype=np.float64).transpose()

df.style.background_gradient(cmap='summer_r',axis=1)
```

```python
#lets check the active cases stateswise

india_cases.sort_values('Active',ascending =False).style\
        .background_gradient(cmap='BrBG_r',subset=['Confirmed'])\
        .background_gradient(cmap='YlGn_r',subset=['Deaths'])\
        .background_gradient(cmap='YlOrBr',subset=['Recovery'])\
        .background_gradient(cmap='BuPu',subset=['Active'])
```

```python
#lets print bar chart

def horizontal_bar_chart(df,x,y,title,x_lable,y_lable,color):
    fig = px.bar(df,x=x,y=y,orientation ='h',title=title,
                labels={x.name:x_lable,
                        y.name:y_lable},color_discrete_sequence=[color])
    fig.update_layout(yaxis={'categoryorder':'total ascending'})
    fig.show()
    
```

```python
top_10_confirmed_Cases = india_cases.sort_values('Confirmed',ascending=False)[:10]
horizontal_bar_chart(top_10_confirmed_Cases,top_10_confirmed_Cases.Confirmed,top_10_confirmed_Cases.index,
                      'Top 10 States with Most confirmed cases','Number of Confirmed cases  (in Thousand)','States name',cnf)
```
